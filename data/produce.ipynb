```python
import os, glob
import pandas as pd
```

```python
df = pd.read_excel("./uc.xlsx")
df["LL"] = df.ID.apply(lambda x: str(x)[:3])

df.LL.unique()
FROM = ['AAR', 'BRU', 'DUB', 'MDC', 'MGP', 'PRG', 'PTO']
TO = ["Aarhus","Brussels","Dublin","Madrid MDC","Madrid MGP","Prague", "Porto"]
for k in range(len(FROM)):
    df.LL = df.LL.apply(lambda x: x.replace(FROM[k],TO[k])) 
df
```

```python
template = """---
layout: default
title: TITLE
parent: PARENT
nav_order: ORDER
---

TXT"""
```

```python
for LL in df.LL.unique():
    if not os.path.exists('../'+LL):
        os.makedirs('../'+LL)
    DF = df[df.LL == LL].reset_index(drop=True)
    for ix, row in DF.iterrows():
        md = template.replace("TITLE",row["ID"]).replace("PARENT",row["LL"]).replace("ORDER",str(ix+1))
        txt = ""
        for col in list(DF.columns[2:-1]):
            txt += "\n\n### "+col+"\n\n"+str(row[col])+"\n\n"
        md = md.replace("TXT",txt)
        with open("../"+LL+"/"+row["ID"]+".md","w") as f:
            f.write(md)
DF.tail(2)
```
